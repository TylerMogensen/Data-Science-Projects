```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('nutrition.csv')
data = data.fillna(0)
```

```python
data['total_fat'] = data['total_fat'].astype(str).replace('g', '', regex=True)
data['total_fat'] = data['total_fat'].astype(float)

data['saturated_fat'] = data['saturated_fat'].astype(str).replace('g', '', regex=True)
data['saturated_fat'] = data['saturated_fat'].astype(float).fillna(0)

data['cholesterol'] = data['cholesterol'].astype(str).replace('mg', '', regex=True)
data['cholesterol'] = data['cholesterol'].astype(float)

data['sodium'] = data['sodium'].astype(str).replace('mg', '', regex=True)
data['sodium'] = data['sodium'].astype(float)

data['vitamin_a'] = data['vitamin_a'].astype(str).replace('mcg', '', regex=True)
data['vitamin_a'] = data['vitamin_a'].astype(str).replace('IU', '', regex=True)
data['vitamin_a'] = data['vitamin_a'].astype(float)

data['vitamin_b'] = data['vitamin_b'].astype(str).replace('mcg', '', regex=True)
data['vitamin_b'] = data['vitamin_b'].astype(str).replace('IU', '', regex=True)
data['vitamin_b'] = data['vitamin_b'].astype(float)


data['vitamin_c'] = data['vitamin_c'].astype(str).replace('mg', '', regex=True)
data['vitamin_c'] = data['vitamin_c'].astype(str).replace('IU', '', regex=True)
data['vitamin_c'] = data['vitamin_c'].astype(float)


data['vitamin_d'] = data['vitamin_d'].astype(str).replace('mg', '', regex=True)
data['vitamin_d'] = data['vitamin_d'].astype(str).replace('IU', '', regex=True)
data['vitamin_d'] = data['vitamin_d'].astype(float)
```

```python
data['calcium'] = data['calcium'].astype(str).replace('mg', '', regex=True)
data['calcium'] = data['calcium'].astype(float)

data['iron'] = data['iron'].astype(str).replace('mg', '', regex=True)
data['iron'] = data['iron'].astype(float)

data['potassium'] = data['potassium'].astype(str).replace('mg', '', regex=True)
data['potassium'] = data['potassium'].astype(float)

data['protein'] = data['protein'].astype(str).replace('g', '', regex=True)
data['protein'] = data['protein'].astype(float)
```

```python
data['carbohydrate'] = data['carbohydrate'].astype(str).replace('g', '', regex=True)
data['carbohydrate'] = data['carbohydrate'].astype(float)

data['fiber'] = data['fiber'].astype(str).replace('g', '', regex=True)
data['fiber'] = data['fiber'].astype(float)

data['sugars'] = data['sugars'].astype(str).replace('g', '', regex=True)
data['sugars'] = data['sugars'].astype(float)

data['fat'] = data['fat'].astype(str).replace('g', '', regex=True)
data['fat'] = data['fat'].astype(float)

data['water'] = data['water'].astype(str).replace('g', '', regex=True)
data['water'] = data['water'].astype(float).fillna(0)
```

```python
data['Protein Category'] = 'None'

data['Protein Category'].loc[data['protein'] >= 1] = 'Low'

data['Protein Category'].loc[data['protein'] >= 10] = 'Moderate'

data['Protein Category'].loc[data['protein'] >= 20] = 'High'
```

```python
data['Glycemic Category'] = 'Low'

data['Glycemic Category'].loc[(data['carbohydrate'] > 25 ) & (data['fiber'] > 10 )] = 'Moderate'

data['Glycemic Category'].loc[(data['carbohydrate'] > 25) & (data['fiber'] < 10 )] = 'High'
```

```python
data['Sodium Category'] = 'Moderately Healthy'

data['Sodium Category'].loc[(data['sodium'] < 100) & (data['cholesterol'] < 10)] = 'Heart Healthy'

data['Sodium Category'].loc[(data['sodium'] > 1000) ] = 'Not Heart Health'
```

```python
data['Calorie Category'] = 'None'

data['Calorie Category'].loc[(data['calories'] < 40)] = 'Low'

data['Calorie Category'].loc[(data['calories'] >= 40)] = 'Moderate'

data['Calorie Category'].loc[(data['calories'] > 400)] = 'High'
```

```python
import sklearn.model_selection as mod_sel
from sklearn.linear_model import LinearRegression
from sklearn import metrics
xtrain, xtest, ytrain, ytest = mod_sel.train_test_split(x, y, test_size = 0.2, random_state = 18)
```

```python
RLR = LinearRegression().fit(xtrain, ytrain)
LRtrainpred = RLR.predict(xtrain)
LRtestpred = RLR.predict(xtest)
```

```python
trainmse = metrics.mean_squared_error(ytrain, LRtrainpred)
testmse = metrics.mean_squared_error(ytest, LRtestpred) 
```

```python
print("RLR train MSE:", trainmse)
print("RLR test MSE:", testmse)
```

```python
trainr2 = metrics.r2_score(ytrain, LRtrainpred)
testr2 = metrics.r2_score(ytest, LRtestpred) 
```

```python
print("RLR train r2:", trainr2)
print("RLR test r2:", testr2)
```

We can see that for the given data that there is likely some overfitting occurring with such a large r2 value. We could probably reduce the number of variables 

```python
from sklearn.feature_selection import SelectKBest, chi2
X_new = SelectKBest(chi2, k=4).fit_transform(x, y)
```
